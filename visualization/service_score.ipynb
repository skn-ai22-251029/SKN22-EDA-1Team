```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/preprocessed_data.csv")
```

```python
# Plot 1: KDE Plot(Hard_Service_Score)
sns.kdeplot(
    data=df, 
    x='Hard_Service_Score', 
    hue='satisfaction', 
    fill=True, 
    common_norm=False, 
    palette='coolwarm', 
)
plt.title('Distribution of Hard Service Score by Satisfaction\n(KDE Plot)')
plt.xlabel('Hard Service Score')
plt.ylabel('Density')

plt.show()

sns.kdeplot(
    data=df, 
    x='Tech_Score', 
    hue='satisfaction', 
    fill=True, 
    common_norm=False, 
    palette='coolwarm', 
)
plt.show()
df['Tech_Score'].value_counts()
```

```python
# order = ['terrible','bad','soso','good','great']
order = ['bad','soso','good','great']
sns.barplot(
    df,
    x='Tech_score_range',
    y = 'satisfaction_score',
    hue= 'Tech_score_range',
    palette='summer',
    order=order
)

plt.show()
```
