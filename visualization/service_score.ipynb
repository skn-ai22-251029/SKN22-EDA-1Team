```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/preprocessed_data.csv")
```

```python
# Plot 1: KDE Plot(Hard_Service_Score)
sns.kdeplot(
    data=df, 
    x='Hard_Service_Score', 
    hue='satisfaction', 
    fill=True, 
    common_norm=False, 
    palette='coolwarm', 
)
plt.title('Distribution of Hard Service Score by Satisfaction\n(KDE Plot)')
plt.xlabel('Hard Service Score')
plt.ylabel('Density')

plt.show()
```

```python
order = ['terrible','bad','soso','good','great']
sns.barplot(
    df,
    x='Tech_score_range',
    y = 'satisfaction_score',
    hue= 'Tech_score_range',
    palette='summer',
    order=order
)

plt.show()
```

```python
order = ['Eco','Eco Plus','Business']
sns.barplot(
    data=df, 
    x='Class', 
    y='satisfaction_score',
    order=order,
    hue= 'Class'
)

plt.show()
```

```python
ct = pd.crosstab(df['Soft_Service_Score'], df['Age_Group'])
sns.heatmap(ct, annot=True, fmt='d', cmap='coolwarm')
```

```python
sns.countplot(data=df, x='Customer Type', hue='Class')
plt.show()
```
